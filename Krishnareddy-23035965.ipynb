```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
data = pd.read_csv('data1.csv')

# Display the first 5 rows Adjust the number according to your preferences in head()
print(data.head())

#This will remove all rows that contain at least one NaN value in any column
data.dropna(inplace=True)

# Check for null value(NAN) if there remove using data.dropna() â€“ Drop all rows that have any NaN values
data.isnull().sum()

# Check for min value of country
data.Country.min()

# Check for mix value of Year name
data.Year.max()
```

```python
# Code for describe data set
data.describe()
```

```python
#A Relational Graph (Line graph) for country ,Product_Category columns.

def line_graph_country_and_Profit(data_line):
    
    #To change the size of graph ---figure() inside figsize() attribute.figsize takes in two parameters â€” width and height.
    plt.figure(figsize=(9, 5))

    #plot graph of 2 columns 'Country' & 'Product_Category'.
    plt.plot(data_line['Country'], data_line['Product_Category'] ,marker="*", label='Country & Product_Category')

    # title of the Graph.
    plt.title('Graph of Country-wise Product_Category', color="skyblue")

    plt.xlabel('Country', color='blue')
    plt.ylabel('Product_Category',color='red')
    
    #display The label name upper left side of the graph.
    plt.grid()
    plt.legend(loc ="upper right")
    plt.show()
    
line_graph_country_and_Profit(data)
```

```python
#Relational Graph(bar graph) for Year wise Occurrences data.

def graph_occurrences_by_Country(data_line):
    
    country_occu = data_line['Country'].value_counts()
    plt.figure(figsize=(8, 4))
    
    # Plotting the bar chart
    plt.bar(country_occu.index, country_occu, label='Country')

    plt.xlabel('Country' ,color="Orange")
    plt.ylabel('Occurrences',color="green")
    plt.legend(loc ="upper right")

    plt.title('Occurrences by Country')
    plt.show()

graph_occurrences_by_Country(data)
```

```python
# A Product_Category(Scatter Graph) explain relationship between Country and profit.

def graph_ountry_profit():
    plt.figure(figsize=(7, 4))

    plt.scatter(data['Country'], data['Profit'], marker = '*', label = 'Country & Profit' ,color='red')
    plt.title('Scatter Graph of Country and Profit', color="blue")

    #setting y sacle as log values
    plt.yscale('log')

    plt.xlabel('COUNTRY', color='blue')
    plt.ylabel('PROFIT',color='red')
    plt.legend(loc ="upper right")
    plt.show()

graph_ountry_profit()
```

```python
#A violin plot(Statistical Graph) for Country and Profit data

def violin_plot_Country_Profit():
    plt.figure(figsize=(8, 4))

    dt = sns.violinplot(x='Country', y='Profit', data=data)

    dt.legend(title='', labels=['Profit & Country'])
    plt.xlabel("Country" ,color="blue")
    plt.ylabel("Profit" ,color="Green")
    plt.title("Statistical Graph of Country and Profit data", color="orange")
    plt.grid()
    plt.show()
    
violin_plot_Country_Profit()
```

```python
#pie chart(Categorical Graph) based on the country Categories occurrences.
def categorical_graph():
    plt.figure(dpi=144)

    country_data = data.groupby('Country').size()

    plt.title("PIE CHART for Country Occurences")
    
    #autopct='%1.3f%%' means display 3 decimal values
    plt.pie(country_data, labels=country_data.index, autopct='%1.3f%%')

    #for the center of circle hole
    center=plt.Circle((0,0),0.5 ,facecolor='skyblue')
    plt.gcf().gca().add_artist(center)

    plt.show()
    
categorical_graph()
```
