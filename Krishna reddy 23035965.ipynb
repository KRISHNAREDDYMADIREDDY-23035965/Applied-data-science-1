```python
import pandas as pd

# Load the dataset
data = pd.read_csv('data1.csv')

# Display the first 5 rows ,if you want more data use value(10) as argument in head()
print(data.head())

# Check for null value(NAN) data
data.isnull().sum()
# Check for min value of country
data.Country.min()
# Check for mix value of country name
data.Country.max()
```

```python
# Code for describe data set
data.describe()
```

```python
#Line graph
import matplotlib.pyplot as plt

#To change the size of graph ---figure() inside figsize() attribute.figsize takes in two parameters — width and height.
plt.figure(figsize=(9, 5))

#plot graph of 2 columns 'Country' & 'Profit'
plt.plot(data['Country'], data['Profit'] ,label='Country & Profit')

# Get the maximum value in the column
max_val = data['Profit'].max()
print(max_val)

# title of the Graph
plt.title('Graph of Year-wise Profit', color="skyblue")

plt.xlabel('Country', color='blue')
plt.ylabel('PROFIT',color='red')
#display The label name upper left of the graph
plt.grid()
plt.legend(loc ="upper right")
plt.show()
```

```python
#Histogram

#ncols=len(data['Year'].unique()): equal to the number of unique years in the 'Year' column of your dataset.

fig, axes = plt.subplots(nrows=1, ncols=len(data['Year'].unique()), figsize=(10, 4))

for (year, val), ax in zip(data.groupby('Year'), axes):
    val.hist(column='Revenue', bins=20, color='skyblue', ax=ax)
    ax.set_xlabel(year)

plt.suptitle('Histogram of Revenue Years-wise')
plt.show()
```

```python
#Scatter Graph
plt.figure(figsize=(7, 4))

plt.scatter(data['Country'], data['Profit'], label='Country & Profit' ,color='red')
plt.title('Graph of Country and Profit', color="skyblue")
plt.yscale('log')
plt.xlabel('Country', color='blue')
plt.ylabel('PROFIT',color='red')
plt.legend(loc ="upper right")
plt.show()
```

```python
# Code for confusion matrix/heatmap/corner/box/violin plot
import seaborn as sns
data2 = pd.read_csv('TSCO_ann.csv')

plt.figure(figsize=(8, 4))
#annot=True display's numeric values of the each cell data 
#fmt=".3f" display's numerical value with 3 decimal places
sns.heatmap(data2.corr(), annot=True, fmt=".3f" )

plt.title('Correlation Heatmap')
plt.show()
```
