```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
data = pd.read_csv('data1.csv')

# Display the first 5 rows ,if you want more data use value(10) as argument in head()
print(data.head())

# Check for null value(NAN) data
data.isnull().sum()
# Check for min value of country
data.Country.min()
# Check for mix value of country name
data.Country.max()
```

```python
# Code for describe data set
data.describe()
```

```python
#Line graph
def line_graph_country_and_Profit(data_line):
    #To change the size of graph ---figure() inside figsize() attribute.figsize takes in two parameters — width and height.
    plt.figure(figsize=(9, 5))

    #plot graph of 2 columns 'Country' & 'Product_Category'
    plt.plot(data_line['Country'], data_line['Product_Category'] ,label='Country & Product_Category')

    # Get the maximum value in the column
    max_val = data_line['Profit'].max()
    print(max_val)

    # title of the Graph
    plt.title('Graph of Country-wise Product_Category', color="skyblue")

    plt.xlabel('Country', color='blue')
    plt.ylabel('Product_Category',color='red')
    #display The label name upper left of the graph
    plt.grid()
    plt.legend(loc ="upper right")
    plt.show()
line_graph_country_and_Profit(data)
```

```python
#Histogram
def histogram_Revenue_and_Year(hist_data): 
    #ncols=len(data['Year'].unique()): equal to the number of unique years in the 'Year' column of your dataset.
    fig, axes = plt.subplots(nrows=1, ncols=len(hist_data['Year'].unique()), figsize=(10, 4))

    for (year, val), ax in zip(hist_data.groupby('Year'), axes):
        val.hist(column='Revenue', bins=20, color='skyblue', ax=ax)
        ax.set_xlabel(year)

    plt.suptitle('Histogram of Revenue Years-wise')
    plt.show()
histogram_Revenue_and_Year(data)
```

```python
#Scatter Graph explain relationship between Country and profit.
plt.figure(figsize=(7, 4))

plt.scatter(data['Country'], data['Profit'], marker = '.', label='Country & Profit' ,color='red')
plt.title('Graph of Country and Profit', color="skyblue")
plt.yscale('log')
plt.xlabel('COUNTRY', color='blue')
plt.ylabel('PROFIT',color='red')
plt.legend(loc ="upper right")
plt.show()
```

```python
# Code for violin plot

plt.figure(figsize=(8, 4))

#annot=True display's numeric values of the each cell data 
#fmt=".3f" display's numerical value with 3 decimal places
#sns.violinplot(x=data2['Rank'], y=data2['Watchtime in Million'])(data2.corr(), annot=True, fmt=".3f" )
sns.violinplot(x='Country', y='Profit', data=data)
plt.title('Violin Plot')
plt.show()
```

```python
#pie chart based on the count of occurrences of each country:

plt.figure(dpi=144)

country_data = data.groupby('Country').size()

plt.title("PIE CHART")
plt.pie(country_data, labels=country_data.index, autopct='%1.3f%%')

#for the center of circle hole
hole=plt.Circle((0,0),0.5 ,facecolor='blue')
plt.gcf().gca().add_artist(hole)

plt.show()
```
